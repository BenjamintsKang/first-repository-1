# 메인 퀘스트 1번 : 은행계좌 만들기

```python
import random as r
# Account클래스
class Account:
  account_count = 0
  def __init__(self,name,money): # __init__ 매서드. 매개변수는 예금주, 초기 잔액
    self.deposit_log = []
    self.withdraw_log = []
    self.put_count = 0
    self.name = name
    self.money = money
    self.bank = 'SC은행'

    num1=""
    for i in range(3):
      num1+=str(r.randint(1,9))
    num2=""
    for i in range(2):
      num2+=str(r.randint(1,9))
    num3=""
    for i in range(6):
      num3+=str(r.randint(1,9))
    self.number = num1+'-'+num2+'-'+num3
    Account.account_count += 1

  def get_account_num(self):
    print(Account.account_count)

  def deposit(self, put):
    if put >= 1:
      self.money += put
      self.deposit_log.append(put)

      self.put_count += 1
      if self.put_count % 5 == 0:
        self.money += self.money*0.01

  def withdraw(self, out):
    if out < self.money:
      self.money -= out
      self.withdraw_log.append(out)

  def display_info(self):
    print("은행이름:",self.bank+", 예금주:",self.name+", 계좌번호:",self.number+", 잔고:",str(format(self.money,','))+"원")

  def deposit_history(self):
    for i in self.deposit_log:
      print(i)

  def withdraw_history(self):
    for i in self.withdraw_log:
      print(i)


customer = []
a = Account("서",10000000)
b = Account("은",20000)
c = Account("재",30000)
customer.append(a)
customer.append(b)
customer.append(c)

for i in customer:
  if i.money > 1000000:
    i.display_info()


a.deposit(100000)
a.deposit(200000)
a.deposit(300000)
a.deposit_history()
```
